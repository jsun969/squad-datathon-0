```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# train = sns.load_dataset("train.csv")
train = pd.read_csv("./train.csv")

train_range = train.copy()
bins = [x / 10.0 for x in range(0, 12, 2)]
labels = [str(x / 10.0) + "-" + str((x + 2) / 10.0) for x in range(0, 10, 2)]
train_range["range"] = pd.cut(
    train_range["grade"], bins=bins, labels=labels, include_lowest=True
)
```

## Family


```python
sns.boxplot(x="mother_job", y="grade", data=train, hue="sex")
plt.title("Mother Job by Grade")
plt.show()
sns.boxplot(x="father_job", y="grade", data=train, hue="sex")
plt.title("Father Job by Grade")
plt.show()
```

## School


```python
# Failure
sns.displot(
    data=train_range, x="range", stat="count", hue="failed_classes", multiple="dodge"
)
plt.title("Grade Influenced by Failure")
plt.show()
```

```python
sns.displot(data=train_range, x="range", stat="count", hue="school_support")
plt.title("Grade Influenced by School Support")
plt.show()
```

## Others


```python
sns.catplot(
    data=train_range,
    x="range",
    y="range",
    estimator=len,
    hue="romantic_rel",
    kind="point",
)
plt.title("Grade Influenced by Romantic Relationship")
plt.xlabel("Range")
plt.ylabel("")

plt.yticks([])

plt.show()
```
