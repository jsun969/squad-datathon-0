```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# train = sns.load_dataset("train.csv")
train = pd.read_csv("./train.csv")
```

## Family


```python
sns.boxplot(x="mother_job", y="grade", data=train, hue="sex")
plt.title("Mother Job by Grade")
plt.show()
sns.boxplot(x="father_job", y="grade", data=train, hue="sex")
plt.title("Father Job by Grade")
plt.show()
```

## School


```python
# Failure
sns.boxplot(x="failed_classes", y="grade", data=train, hue="sex")
plt.title("Grade Influence by Failed Classes")
plt.show()
```
