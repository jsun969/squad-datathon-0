```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# train = sns.load_dataset("train.csv")
train = pd.read_csv("./train.csv")
```

## Family


```python
sns.boxplot(x="mother_job", y="grade", data=train, hue="sex")
plt.title("Mother Job by Grade")
plt.show()
sns.boxplot(x="father_job", y="grade", data=train, hue="sex")
plt.title("Father Job by Grade")
plt.show()
```

## School


```python
# Failure
data = train.copy()
bins = [x / 10.0 for x in range(0, 12, 2)]
labels = [str(x / 10.0) + "-" + str((x + 2) / 10.0) for x in range(0, 10, 2)]
data["range"] = pd.cut(data["grade"], bins=bins, labels=labels, include_lowest=True)
sns.displot(data, x="range", stat="count", hue="failed_classes", multiple="dodge")
plt.title("Grade Influenced by Failure")
plt.show()
```

```python
sns.boxplot(x="study_time", y="grade", data=train)
plt.title("Grade Influence by Study Time")
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Box plot
school_support_categories = ["no", "yes"]
grade_data = [
    train[train["school_support"] == category]["grade"]
    for category in school_support_categories
]

axs[0].boxplot(grade_data, labels=school_support_categories)
axs[0].set_title("Grade Distribution by School Support (Box Plot)")
axs[0].set_xlabel("School Support")
axs[0].set_ylabel("Grade")

# Violin plot
axs[1].violinplot(grade_data)
axs[1].set_xticks([1, 2])
axs[1].set_xticklabels(school_support_categories)
axs[1].set_title("Grade Distribution by School Support (Violin Plot)")
axs[1].set_xlabel("School Support")
axs[1].set_ylabel("Grade")

fig.tight_layout()
plt.show()
```
