```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# train = sns.load_dataset("train.csv")
train = pd.read_csv("./train.csv")
```

## Family


```python
# Mother's Job
sns.catplot(data=train, x="mother_job", y="grade", hue="sex", kind="box")
# Father's Job
sns.catplot(data=train, x="father_job", y="grade", hue="sex", kind="box")
```

## School


```python
# Failure
sns.boxplot(x="failed_classes", y="grade", data=train, hue="sex")
```
